# Bootstrapping

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class SwapRate:
    def __init__(self, tenor, rate):
        self._tenor = tenor
        self._rate = rate
    def __str__(self):
        return f"""{self._tenor} - {100*self._rate}%"""
    def __repr__(self):
        return self.__str__()

class SwapRateCurve:
    def __init__(self, tenors, rates):
        self._tenors = tenors
        self._rates = rates
        self._swap_rates = {
            tenor: SwapRate(tenor, rate) for tenor, rate in zip(tenors, rates)
        }

    def plot_curve(self, grid=True):
        plt.plot(self._tenors, self._rates)
        plt.grid(grid)
```

```python
swap_rate_curve = SwapRateCurve([1, 2, 3, 4, 5], [1.5/100, 1.6/100, 1.7/100, 1.8/100, 1.9/100])
```

```python
swap_rate_curve.plot_curve()
```

```python
def bootstrap_zc(swap_rate_curve:SwapRateCurve):
    tenors, rates = swap_rate_curve._tenors, swap_rate_curve._rates
    zc = []
    for idx, (tenor, rate) in enumerate(zip(tenors, rates)):
        if idx == 0:
            zc.append((1/rate)**(1/tenor) - 1)

    return zc
```
